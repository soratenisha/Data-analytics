columns?


```python
print(df[['Ship Date', 'Order Date', 'Order Quantity']].isnull().sum())

print(df[['Total', 'Profit Margin', 'Discount %']].head())
```

5. Convert Order Date and Ship Date to datetime.


```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

print(df[['Order Date', 'Ship Date']].dtypes)
```

6. Check for future or inconsistent shipping dates.


```python
print((df['Ship Date'] < df['Order Date']).sum())
```

7. Convert price columns to numeric (remove $ and commas).


```python
cols = ['Cost Price', 'Retail Price', 'Profit Margin', 'Sub Total', 'Discount $', 'Order Total', 'Shipping Cost', 'Total']

df[cols] = df[cols].replace('[\$,]', '', regex=True).astype(float)
```

8. What are the unique values in Customer Type and Order Priority?

```python
print(df['Customer Type'].unique())
print(df['Order Priority'].unique())
```

9. What are the most common shipping modes?

```python
print(df['Ship Mode'].value_counts())
```

10. Which cities have the highest number of orders?


```python
print(df['City'].value_counts().head())
```

11. What’s the range of order quantities and prices?

```python
df['Total'] = df['Total'].replace('[\$,]', '', regex=True).astype(float)

print(df[['Order Quantity', 'Total']].agg(['min', 'max']))
```

12. Create a new column for shipping duration.


```python
df['Shipping Duration'] = (pd.to_datetime(df['Ship Date']) - pd.to_datetime(df['Order Date'])).dt.days

print(df[['Order Date', 'Ship Date', 'Shipping Duration']].head())
```

13. Are there any orders with zero or negative total or quantity?

```python
print((df['Order Quantity'] <= 0).sum())
print((df['Total'] <= 0).sum())
```


14. Are all discount percentages matching discount dollar amounts?

```python
sub_total = df['Sub Total'].replace('[\$,]', '', regex=True).astype(float)
disc_pct = df['Discount %'].str.rstrip('%').astype(float) / 100
disc_dollar = df['Discount $'].replace('[\$,]', '', regex=True).astype(float)

mismatch_count = (round(sub_total * disc_pct, 2) != disc_dollar).sum()

print(f"Total Mismatches: {mismatch_count}")
```



15. Check for mismatches in total calculation’

```python

cols = ['Sub Total', 'Discount $', 'Shipping Cost', 'Total']
df[cols] = df[cols].replace('[\$,]', '', regex=True).astype(float)

mismatches = (df['Total'] != (df['Sub Total'] - df['Discount $'] + df['Shipping Cost'])).sum()
print(f"Total Mismatches: {mismatches}")
```



16. Identify top 5 products by order quantity.
