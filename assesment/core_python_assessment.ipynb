```python
import json
import os

class ClinicAppointment:
    def __init__(self):
        self.slots = ["10am", "11am", "12pm", "2pm", "3pm"]

        if os.path.exists("appointments.json"):
            with open("appointments.json", "r") as f:
                self.appointments = json.load(f)
        else:
            self.appointments = []

    def save_data(self):
        with open("appointments.json", "w") as f:
            json.dump(self.appointments, f)

    def book_appointment(self):
        name = input("Patient Name: ") 
        age = input("Age: ")
        mobile = input("Mobile Number: ")
        doctor = input("Doctor Name: ")

        print("Available Time Slots:")
        for s in self.slots:
            print(s)

        slot = input("Choose Slot: ")

        if slot not in self.slots:
            print("Invalid slot!")
            return

        count = 0
        for appt in self.appointments:
            if appt["doctor"] == doctor and appt["slot"] == slot:
                count += 1

        if count >= 3:
            print("Slot full for this doctor!")
            return

        self.appointments.append({
            "name": name,
            "age": age,
            "mobile": mobile,
            "doctor": doctor,
            "slot": slot
        })

        self.save_data()
        print("Appointment Booked Successfully")

    def view_appointment(self):
        mobile = input("Enter Mobile Number: ")
        found = False

        for appt in self.appointments:
            if appt["mobile"] == mobile:
                print("\nAppointment Details")
                print("Name:", appt["name"])
                print("Doctor:", appt["doctor"])
                print("Slot:", appt["slot"])
                found = True

        if not found:
            print("No appointment found")

    def cancel_appointment(self):
        mobile = input("Enter Mobile Number: ")

        for appt in self.appointments:
            if appt["mobile"] == mobile:
                self.appointments.remove(appt)
                self.save_data()
                print("Appointment Cancelled")
                return

        print("No appointment found")

    def menu(self):
        print("\n1. Book Appointment")
        print("2. View Appointment")
        print("3. Cancel Appointment")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            self.book_appointment()
        elif choice == "2":
            self.view_appointment()
        elif choice == "3":
            self.cancel_appointment()
        elif choice == "4":
            print("Thank You!")
        else:
            print("Invalid Choice!")

clinic = ClinicAppointment()
clinic.menu()
```

```python
import json, os

class BusReservation:

    def __init__(self):
        self.routes = {
            "Mumbai to Pune": 500,
            "Delhi to Jaipur": 600,
            "Bangalore to Chennai": 550,
            "Hyderabad to Goa": 700
        }

        if os.path.exists("bus.json"):
            with open("bus.json","r") as f:
                self.bookings = json.load(f)
        else:
            self.bookings = {}

    def save(self):
        with open("bus.json","w") as f:
            json.dump(self.bookings,f)

    def show_routes(self):
        print("\nRoutes:")
        routes = list(self.routes.keys())
        for i,r in enumerate(routes,1):
            print(i,".",r,"- ₹",self.routes[r])
        return routes

    def book_ticket(self):
        name=input("Name: ")
        mobile=input("Mobile: ")

        routes=self.show_routes()
        c=int(input("Select route number: "))
        route=routes[c-1]

        ticket_id="T"+str(len(self.bookings)+1)

        self.bookings[ticket_id]={
            "name":name,
            "mobile":mobile,
            "route":route
        }

        self.save()

        print("Ticket Booked! ID:",ticket_id)

    def view_ticket(self):
        tid=input("Enter Ticket ID: ")

        if tid in self.bookings:
            t=self.bookings[tid]
            print("Name:",t["name"])
            print("Route:",t["route"])
        else:
            print("Ticket not found")

    def cancel_ticket(self):
        tid=input("Enter Ticket ID: ")

        if tid in self.bookings:
            del self.bookings[tid]
            self.save()
            print("Ticket Cancelled")
        else:
            print("Invalid Ticket ID")


system=BusReservation()

print("\n1 Show Routes")
print("2 Book Ticket")
print("3 View Ticket")
print("4 Cancel Ticket")

choice=input("Enter choice: ")

if choice=="1":
    system.show_routes()

elif choice=="2":
    system.book_ticket()

elif choice=="3":
    system.view_ticket()

elif choice=="4":
    system.cancel_ticket()
```

```python
import json
import os

class SchoolManagement:

    def __init__(self):
        if os.path.exists("students.json"):
            with open("students.json", "r") as f:
                data = json.load(f)
                self.students = data["students"]
                self.id = data["id"]
        else:
            self.students = []
            self.id = 1

    def save(self):
        with open("students.json", "w") as f:
            json.dump({"students": self.students, "id": self.id}, f)

    def new_admission(self):

        name = input("Name: ")
        age = int(input("Age: "))
        mobile = input("Mobile (10 digits): ")

        if age < 5 or age > 18:
            print("Invalid Age")
            return

        if len(mobile) != 10:
            print("Invalid Mobile")
            return

        student = {
            "id": self.id,
            "name": name,
            "age": age,
            "mobile": mobile
        }

        self.students.append(student)

        print("Admission Done")
        print("Student ID:", self.id)

        self.id += 1
        self.save()

    def view_student(self):

        sid = int(input("Enter Student ID: "))

        for s in self.students:
            if s["id"] == sid:
                print("Name:", s["name"])
                print("Age:", s["age"])
                print("Mobile:", s["mobile"])
                return

        print("Student Not Found")

    def remove_student(self):

        sid = int(input("Enter Student ID: "))

        for s in self.students:
            if s["id"] == sid:
                self.students.remove(s)
                self.save()
                print("Record Removed")
                return

        print("Student Not Found")


school = SchoolManagement()

print("\n1. New Admission")
print("2. View Student")
print("3. Remove Student")
print("4. Exit")

ch = input("Enter Choice: ")

if ch == "1":
    school.new_admission()

elif ch == "2":
    school.view_student()

elif ch == "3":
    school.remove_student()

elif ch == "4":
    print("Exit ")

else:
    print("Invalid Choice")
```
